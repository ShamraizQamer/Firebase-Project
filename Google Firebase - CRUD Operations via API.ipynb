# Insert 1 record in Firebase

```python
import requests
import pandas as pd
from requests.structures import CaseInsensitiveDict
from pytz import timezone
from datetime import datetime
import json

url = "<Enter firebase realtime database URL>/learn/master_group/14.json"


headers = CaseInsensitiveDict()
headers["Content-Type"] = "application/json"

data = {"count": "0", "group" : "No-Code App Dev", "id" : 14}
data = json.dumps(data, indent = 4) 

resp = requests.put(url, headers=headers, data=data)

print(resp.status_code)
print(resp.reason)
print(resp.json())

myjson = (resp.json())


#df = pd.DataFrame(myjson)
#df.head()
```

# Update a record in Firebase

```python
import requests
import pandas as pd
from requests.structures import CaseInsensitiveDict
from pytz import timezone
from datetime import datetime
import json

url = "<Enter firebase realtime database URL>/learn/master_group/14.json"



headers = CaseInsensitiveDict()
headers["Content-Type"] = "application/json"

data = {"group" : "Firebase & AppGyver"}
data = json.dumps(data, indent = 4) 

resp = requests.patch(url, headers=headers, data=data)

print(resp.status_code)
print(resp.reason)
print(resp.json())

myjson = (resp.json())


#df = pd.DataFrame(myjson)
#df.head()
```
